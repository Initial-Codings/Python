```python
# Corey 
# May / 17 / 2017
# my follow thorough code
# class variables: variables that are shared amongst
# are shared among all instances of a class
# class varibles should be same for each instance
# Example he used annual raise

# While instance variables can be unique for each 
# instance like email, names and pay 

# Order of checking for varibles
# 1 instance(namepace)
# 2 Class then it inherits from contains that attribute(namespace)

# Number of employees should be the same for all instances
class Employee:
    '''Represent an Employee'''
    # Class variables /atrributes
    raise_amount = 1.04
    num_of_emps  = 0
    
    # Characterists and Instanciate Employee Class
    def __init__(self, first, last, pay):
        '''
        Initialize Employee class object
        INPUT: string(first and last) int(pay)  
        '''
        # Attributes (4)
        self.first = first
        self.last  = last
        self.pay   = pay
        self.email = first + '.' + last + '@company.com'
        
        # Incremented here because each employee's instance is done here
        Employee.num_of_emps += 1
        
    # Methods(2)    
    def fullname(self):
        '''
        Displays Employee's full name
        INPUT: class instance
        '''
        return '{} {}'.format(self.first, self.last)
    
    def apply_raise(self):
        '''
        Return the raise amount of an Employee
        INPUT: class instance
        '''
        self.pay = int(self.pay * self.raise_amount)   
    
# Employee Instanciation
emp1 = Employee('Corey','Schafer', 50000)
emp2 = Employee('Test','User', 60000)

print 'Class Attribute:{}Raise amount {}{}'.format('\n', Employee.raise_amount,'\n')
emp1.raise_amount # inherited from class thus same as class
emp2.raise_amount # inherited from class thus same as class

## This are further summary from the video

# This are accessing the class raise_amount
#print emp1.raise_amount
#print emp1.raise_amount

# To get a better idea of what's happening seeing the
# Namespace of instances which is a dictionary variable of the instance

# Special trick to see the name space

# To see atributes of particular instances 
print "To see the name space and Class Composition"
print emp1.__dict__
print emp2.__dict__

# To see the class composition(pointers) and attributes
# raise_amount only in this name space
print Employee.__dict__

# self on the apply_raise method apply facilitates this
# If used Employee.raise_amount instead this wont work
# Example when not to use self for example keeping track of the # of Employees
print ''
emp1.raise_amount = 1.05

# Notice raise_amount now in the name space of emp1
print 'emp1 New Name Space'
print emp1.__dict__

# As seen here number of employess in this instance they are two
print ''
print 'Total number of employees in this instance'
print Employee.num_of_emps
```
