```python
# Previous code
%matplotlib inline
m = Basemap(projection = 'mill')
m.drawcoastlines() # coastol lines of the contients
plt.show()
```

```python
# How to zoom in 
%matplotlib inline
m = Basemap(projection = 'mill',
           llcrnrlat = -40,
           llcrnrlon = -40,
           urcrnrlat = 50,
           urcrnrlon = 75) # zoom to Africa
m.drawcoastlines()
plt.show()
```

```python
%matplotlib inline
m = Basemap(projection='mill',
            llcrnrlat = -90,
            llcrnrlon = -180,
            urcrnrlat = 90,
            urcrnrlon = 180) # zoomed out to see the word

m.drawcoastlines()
m.drawcountries(linewidth=2) # countries of the world
m.drawstates(color='b')      # state boundries in North America
m.drawcounties(color='darkred') # counties in USA

plt.title('Basemap Tutorial') # Map Title
plt.show()
```

```python
%matplotlib inline
m = Basemap(projection='mill',
            llcrnrlat = -90,
            llcrnrlon = -180,
            urcrnrlat = 90,
            urcrnrlon = 180,
           resolution = 'l')
m.drawcoastlines()
plt.show()
```

```python
# Didnt result like his
%matplotlib inline
m = Basemap(projection='mill',
            llcrnrlat = -90,
            llcrnrlon = -180,
            urcrnrlat = 90,
            urcrnrlon = 180,
           resolution = 'l')
m.drawcoastlines()
m.etopo()
plt.show()
```
