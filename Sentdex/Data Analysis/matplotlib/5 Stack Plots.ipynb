```python
# Wholistic day
days =     [1,2,3,4,5]

# Parts that make up the day
# Each column adds up to 24 hrs total wholistic hour for a day
sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

#stack plot
# plt.stackplot(x, *args, **kwargs)
plt.stackplot(days, sleeping, eating, working, playing, 
              colors=['m','c','r','k'])

# Label
plt.xlabel('x')
plt.ylabel('y')

# Title 
plt.title('Interesting Graph\nCheck it out')

# Display Graph
plt.show()

# labels for Stack plot
# Everytime you have a fill matplotlib term(polygons) cant
# have a label
```

```python
# Can go around making fake plots
# Variables (data)
days     = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating   = [2,3,4,3,2]
working  = [7,8,7,2,2]
playing  = [8,5,7,8,13]

# fake plots empty list to solve the label issue
plt.plot([], [], color = 'm', label = 'Sleeping', linewidth=5)
plt.plot([], [], color = 'c', label = 'Eating',   linewidth=5)
plt.plot([], [], color = 'r', label = 'Working',  linewidth=5)
plt.plot([], [], color = 'k', label = 'Playing',  linewidth=5)

#stack plot
plt.stackplot(days, sleeping, eating, working, playing, colors = ['m','c','r','k'])

# Label
plt.xlabel('x')
plt.ylabel('y')

# Title and legend
plt.title('Interesting Graph\nCheck it out')
plt.legend()

# Display Graph
plt.show()

# Now we have a label
```
