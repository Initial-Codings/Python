```python
# Part 1
# Variables
x = []
y = []

# Used csv but we can do this another way
# Used csv because it is built in
with open('data for matplotlib.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
# Chart
plt.plot(x, y, label = 'Loaded from file!')

# Label, title, and legend
plt.xlabel('{}'.format('x'))
plt.ylabel('{}'.format('y'))
plt.title('{}'.format('Sample Graph\nHey am down here'))
plt.legend()

# Display
plt.show()
```

```python
# Part 2
# Another way
# More popular way

# Variables
x, y = np.loadtxt('data for matplotlib.txt', delimiter=',', unpack = True)

# Chart
plt.plot(x, y, label = 'Loaded from file!')

# Label, title, and legend
plt.xlabel('{}'.format('x'))
plt.ylabel('{}'.format('y'))
plt.title('{}'.format('Sample Graph\nHey am down here'))
plt.legend()

# Display
plt.show()
```
