# add_subplot
___

```python
%matplotlib inline
# data 
# Need to here to plot as desired
fig = plt.figure()
style.use('fivethirtyeight')

def create_plots():
    xs = []
    ys = []
    
    for i in xrange(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
    return xs, ys

# Defining the figure
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

# Graphing the fig
ax1.plot(xs, ys)
ax2.plot(xs, ys)
# title for the subplot 
ax1.set_title('ax1')
ax2.set_title('ax2')
# make them neater
fig.tight_layout()
plt.show()
```

```python
%matplotlib inline
# data 
# Need to here to plot as desired
fig = plt.figure()
style.use('fivethirtyeight')

def create_plots():
    xs = []
    ys = []
    
    for i in xrange(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
    return xs, ys

# Defining the figure
ax1 = plt.subplot2grid((2,2), (0,0), rowspan = 1,colspan = 1)
ax2 = plt.subplot2grid((2,2), (0,1), rowspan = 1,colspan = 1)
ax3 = plt.subplot2grid((2,2), (1,0), rowspan = 1,colspan = 2)

# Graphing the fig
ax1.plot(xs, ys)
ax2.plot(xs, ys)
ax3.plot(xs, ys)
# title for the subplot 
ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
# make them neater
fig.tight_layout()
plt.show()
```
