```python
# 1. Was to plot the points which were NYC and LA

%matplotlib inline
# map
m = Basemap(projection='mill',
            llcrnrlat = 25,
            llcrnrlon = -130,
            urcrnrlat = 50,
            urcrnrlon = -60,
            resolution='l')

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')

# points coordinates to draw on the map

# New york our destination
NYClat, NYClon = (40.7127, -74.0059) # New York # (lat, long)
xpt, ypt = m(NYClon, NYClat) # now (xpoint, ypoint)                                       
m.plot(xpt, ypt, 'r*', markersize = 10) # to plot point

#LA our origin
LAlat, LAlon = (34.05, -118.25) # Los Angeles # (lat, long)
xpt, ypt = m(LAlon, LAlat) # now (xpoint, ypoint)                                       
m.plot(xpt, ypt, 'g^', markersize = 10) # to plot point
```

```python
#2 To plot the line and arc using list

%matplotlib inline
# map
m = Basemap(projection='mill',
            llcrnrlat = 25,
            llcrnrlon = -130,
            urcrnrlat = 50,
            urcrnrlon = -60,
            resolution='l')

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')

# points coordinates to draw on the map

xs = []
ys = []

# New york our destination
NYClat, NYClon = (40.7127, -74.0059) 
xpt, ypt = m(NYClon, NYClat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt, ypt, 'r*', markersize = 10) 

#LA our origin
LAlat, LAlon = (34.05, -118.25) 
xpt, ypt = m(LAlon, LAlat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt, ypt, 'g^', markersize = 10) 

# plot the whole thing
m.plot(xs, ys, 'b--', linewidth = 2.5, label = 'LA to NYC')
m.drawgreatcircle(NYClon, NYClat, LAlon, LAlat, color='c', linewidth=2.5, label='Arc')

plt.legend(loc = 4)
plt.title('Basemap')
```
